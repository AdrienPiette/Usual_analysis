```python
df_2.drop(columns=['Site','Fracture ID','Artifact ID'], inplace=True)
```

```python
object_cols = [col for col in df_2.select_dtypes(include='object').columns if col != 'Artifact ID']

for col in object_cols:
    if df_2[col].dtype == 'object':
        df_2[col] = df_2[col].astype('category')
    else:
        df_2[col] = pd.to_numeric(df_2[col], errors='coerce')

df_2
```

```python
df.columns
```

```python
common_cols = set(df.columns).intersection(df_2.columns)
print("Colonnes en commun :", list(common_cols))
```

```python
cols_to_keep = list(common_cols) + ['Mode of Propulsion', 'attribute group']

# Prepare df_2 with the same columns, adding 'Mode of Propulsion' and 'attribute group' as NaN
df_2_common = df_2[list(common_cols)].copy()
df_2_common['Mode of Propulsion'] = np.nan
df_2_common['attribute group'] = np.nan

combined_common = pd.concat([df[cols_to_keep], df_2_common], ignore_index=True)
combined_common
```

```python
combined_common = combined_common.dropna(subset=['Mode of Propulsion'])
combined_common
```

```python
data_training = combined_common.copy()

data_training.to_excel(r'C:\Users\pieta\OneDrive\Bureau\tracéo_justin\Usual_analysis\data\data_training.xlsx', index=False)
```
